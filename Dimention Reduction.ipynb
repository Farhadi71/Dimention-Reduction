# LDA

```python
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
df['target'] = cancer['target']

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101 )
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis()
lda.fit(x_train, y_train)
```

```python
lda.explained_variance_ratio_
```

```python
lda = LinearDiscriminantAnalysis(n_components=1)
x_lda_train = lda.fit_transform(x_train, y_train)
x_lda_test = lda.transform(x_test)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
grid_param = [
    {
        'C': [0.1, 1, 10, 100, 1000],
        'gamma': [0.0001, 0.001, 0.01, 0.1, 1],
        'kernel':['rbf'],
    },
    {
        'C': [0.1, 1, 10, 100, 1000],
        'kernel':['linear']
    },
]

grid = GridSearchCV(SVC(), grid_param, verbose=2)
grid.fit(x_lda_train, y_train)
```

```python
y_hat = grid.predict(x_lda_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_hat, y_test))
```

```python
print(classification_report(y_hat, y_test))
```

# LDA with wine dataset

```python
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
wine.keys()
```

```python
df = pd.DataFrame(wine['data'], columns=wine['feature_names'])
df['target'] = wine['target']

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101 )
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis()
lda.fit(x_train, y_train)
```

```python
lda.explained_variance_ratio_
```

```python
lda = LinearDiscriminantAnalysis(n_components=2)
x_lda_train = lda.fit_transform(x_train, y_train)
x_lda_test = lda.transform(x_test)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
grid_param = [
    {
        'C': [0.1, 1, 10, 100, 1000],
        'gamma': [0.0001, 0.001, 0.01, 0.1, 1],
        'kernel':['rbf'],
    },
    {
        'C': [0.1, 1, 10, 100, 1000],
        'kernel':['linear']
    },
]

grid = GridSearchCV(SVC(), grid_param, verbose=2)
grid.fit(x_lda_train, y_train)
```

```python
y_hat = grid.predict(x_lda_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_hat, y_test))
print(classification_report(y_hat, y_test))
```

# T-SNE

```python
from sklearn.datasets import fetch_openml
```

```python
mnist = fetch_openml('mnist_784', version=1, parser='auto')
mnist_target = mnist.target.astype(np.uint8)
```

```python
mnist.keys()
```

```python
X = mnist['data']
y = mnist['target'] 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.95, random_state=101)
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, perplexity=100, init='random', random_state=42)
x_tsne = tsne.fit_transform(X_train)
plt.scatter(x_tsne[:, 0], x_tsne[:, 1], c = y_train.astype(np.uint8))
```

```python
tsne = TSNE(n_components=2, perplexity=100, metric='correlation', init='pca')
x_tsne = tsne.fit_transform(X_train)
plt.scatter(x_tsne[:, 0], x_tsne[:, 1], c = y_train.astype(np.uint8))
```

```python
classes = np.unique(y_train)

for i in range(1, 202,50):
    fig, (plt1, plt2, plt3) = plt.subplots(1, 3, figsize=(13, 4))
    tsne = TSNE(n_components=2, perplexity=i, init='random' )
    x_tsne = tsne.fit_transform(X_train)
    s1 = plt1.scatter(x_tsne[:, 0], x_tsne[:, 1], c = y_train.astype(np.uint8), cmap = 'viridis')

    tsne = TSNE(n_components=2, perplexity=i, metric='correlation', init='pca')
    x_tsne = tsne.fit_transform(X_train)
    s2 = plt2.scatter(x_tsne[:, 0], x_tsne[:, 1], c = y_train.astype(np.uint8), cmap = 'viridis')


    tsne = TSNE(n_components=2, perplexity=i, init='pca')
    x_tsne = tsne.fit_transform(X_train)
    s3 = plt3.scatter(x_tsne[:, 0], x_tsne[:, 1], c = y_train.astype(np.uint8), cmap = 'viridis')
    plt.show()
```
